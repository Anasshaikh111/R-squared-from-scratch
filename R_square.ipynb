```python
import matplotlib.pyplot as plt
```

```python


# Step 1: Function to calculate R from raw X, Y
def r_squared_from_data(x, y):
    n = len(x)
    
    # Calculate means
    mean_x = sum(x) / n
    mean_y = sum(y) / n

    # Calculate slope (m) and intercept (c) of best-fit line
    num = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n))
    den = sum((x[i] - mean_x) ** 2 for i in range(n))
    m = num / den
    c = mean_y - m * mean_x

    # Predicted y values
    y_pred = [m * xi + c for xi in x]

    # Compute R
    ss_tot = sum((yi - mean_y) ** 2 for yi in y)
    ss_res = sum((y[i] - y_pred[i]) ** 2 for i in range(n))
    r2 = 1 - (ss_res / ss_tot)

    return r2, y_pred, m, c


```

```python
# Step 2: Input Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
```

```python
# Step 3: Calculate R and predicted values
r2, y_pred, slope, intercept = r_squared_from_data(x, y)
print("R_squar:", r2)
```

```python
# Step 4: Plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label="Actual data")
plt.plot(x, y_pred, color='red', label=f"Regression line\ny = {slope:.2f}x + {intercept:.2f}")
plt.title(f"Linear Regression Fit\nR = {r2:.3f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
